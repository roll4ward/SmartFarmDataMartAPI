### 4. 항목코드 (FatrCode)

        - 이를 위해 fatrcode 가 어떤 분류 코드에 속하는지에 대한 함수를 제공함.
        - `sectcode_of_fatrcode(fatrcode: str)`

```python
from codes.appendix import FatrCode

# CC01_1_B_C 가 어떤 분류 코드에 속하는지 return 하는 함수
FatrCode.sectcode_of_fatrcode("CC01_1_B_C")
```
